```python
import coalestr as cs
import matplotlib.pyplot as plt
```

```python
my_species = cs.species()
my_species.plot_history('Nh')
```

```python
my_species.get_coalescent()
my_species.get_diversity()
```

```python
my_region = cs.region(metapopulation = my_species)
my_region.get_coalescent()
my_region.get_diversity()
```

```python
my_history = [[1000, 20, 3, 0.1, 3],
              [20, 80, 3, 1, 3],
              [50, 20, 3, 0.1, 3]]
my_village = cs.Population(my_history, metapopulation = my_region)
```

```python
observe = range(100)
my_village.get_coalescent(observe, show = False)
```

```python
my_village.get_diversity(show = False)
my_village.plot_observations()
```
