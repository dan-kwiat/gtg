```python
import coalestr as cs
import matplotlib.pyplot as plt
```

```python
metapop_history = [[100000, 3000, 10, 1, 0]]
metapop = cs.Population(metapop_history)
metapop.get_coalescent()
metapop.get_diversity()
```

```python
subpop_history = [[300, 10, 5, 0, 1],
                  [50, 30, 5, 1, 1],
                  [50, 10, 5, 0, 1]]

observations = range(0, 150, 1)

subpop = cs.Population(subpop_history, metapopulation = metapop)
subpop.get_coalescent(observations, show = False)
subpop.get_diversity(show = False)
subpop.plot_observations(metrics = ("N", "X", "snp_het", "hap_hom"))
```

```python
subpop_history = [[300, 10, 5, 0.5, 1],
                  [50, 30, 5, 0.5, 1],
                  [50, 10, 5, 0.5, 1]]

observations = range(0, 150, 1)

subpop = cs.Population(subpop_history, metapopulation = metapop)
subpop.get_coalescent(observations, show = False)
subpop.get_diversity(show = False)
subpop.plot_observations(metrics = ("N", "snp_het", "hap_hom"))
```
