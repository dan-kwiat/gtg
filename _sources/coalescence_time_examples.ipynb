This code uses the coalestr.py module to explore the probability distribution of time to coalescence of two alleles, for different transmission parameters.

```python
import coalestr as cs
import matplotlib.pyplot as plt
```

```python
history = [[10000, 100, 1, 0, 0]]
pop_1 = cs.Population(history)
pop_1.get_coalescent(show = True)
```

```python
history = [[10000, 100, 2, 0.5, 0]]
pop_2 = cs.Population(history)
pop_2.get_coalescent(show = True)
```

```python
history = [[10000, 100, 10, 1, 0]]
pop_3 = cs.Population(history)
pop_3.get_coalescent(show = True)
```

```python
## SET UP THE PLOTS

plt.figure(figsize=(12,4))

ax1 = plt.subplot(1,2,1)
ax1.set_xlabel("Time to coalescence", fontsize=12)
ax1.set_ylabel("Probability", fontsize=12)
ax1.set_title("Between-host", position=(0.8, 0.3), fontsize=14)
ax1.set_xscale('log',basex=10)

ax2 = plt.subplot(1,2,2)
ax2.set_xlabel("Time to coalescence", fontsize=12)
ax2.set_ylabel("Probability", fontsize=12)
ax2.set_title("Within-host", position=(0.8, 0.3), fontsize=14)
ax2.set_xscale('log',basex=10)

## RUN THE PLOTS

for pop in [pop_1, pop_2, pop_3]:

## PLOT OUTPUT

    ax1.plot(range(1, pop.t_sim), pop.coalescent[0, 1:, 0], marker='', linewidth=2, label=
        "N {0:.0f}, Q {1:.0f}, \u03C7 {2:.1f}".format(
        pop.parameters[0,0],
        pop.parameters[0,1],
        pop.parameters[0,2] ))
    
    ax1.legend(frameon=False, fontsize=12)
    
    ax2.plot(range(1, pop.t_sim), pop.coalescent[0, 1:, 1], marker='', linewidth=2, label=
        "N {0:.0f}, Q {1:.0f}, \u03C7 {2:.1f}".format(
        pop.parameters[0,0],
        pop.parameters[0,1],
        pop.parameters[0,2] ))
    
    ax2.legend(frameon=False, fontsize=12)
    
plt.show()
```
